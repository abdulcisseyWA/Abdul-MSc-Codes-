```python
# Calculate the centroid for each building
Building_df['centroid'] = Building_df.geometry.centroid

# Perform the spatial join
joined_df = gpd.sjoin(grid_df, Building_df, how='left', op='contains')


# Now grid_df has an additional column 'building_count' with the number of buildings per grid cell
joined_df.head()
```

```python
for col in joined_df.columns:
    print(col)
```

```python
joined_df.describe()
```

```python
joined_df.shape
```

```python
# Count the number of buildings per grid cell
building_counts = joined_df.groupby('TARGET_FID').size()

# Add the counts to the original grid dataframe
grid_df['building_count'] = grid_df.index.map(building_counts)

# If there are any NaN values (grids with no buildings), replace them with 0
grid_df['building_count'].fillna(0, inplace=True)

grid_df.head()
```

```python
grid_df.describe()
```

```python
# Calculate the mean of the building attributes for each grid cell
mean_attributes = joined_df.drop(columns=['geometry', 'centroid']).groupby('TARGET_FID').mean()

# Add a suffix to the mean attributes columns
mean_attributes = mean_attributes.add_suffix('_m')

# Add the counts and mean attributes to the original grid dataframe
grid_df = grid_df.join(mean_attributes, on='TARGET_FID')

grid_df.head()
```
